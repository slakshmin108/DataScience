```
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
from sklearn.metrics.pairwise import cosine_similarity
```

```
# URL for BERT models
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"

# instantiate the BERT models
bert_preprocess_model = hub.KerasLayer(preprocess_url)
bert_model = hub.KerasLayer(encoder_url)
```

```
# sentences to be compared
key_phrases = ['he is happy', 'he is sad']

# preprocess the sentences
text_preprocessed = bert_preprocess_model(key_phrases)
text_preprocessed.keys()
```

```
# generate sentence embeddings using BERT
bert_results = bert_model(text_preprocessed)
bert_results.keys()
```

```
sentence_embeddings = bert_results['pooled_output']
sentence_embeddings
```

```
# calculate the cosine similarity between the sentences
cosine_similarity([sentence_embeddings[0]], [sentence_embeddings[1]])
```
